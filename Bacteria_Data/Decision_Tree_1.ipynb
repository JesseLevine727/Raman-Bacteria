```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import classification_report, accuracy_score


import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Y.csv")
```

```python
spectra = data.iloc[:, :-1]
label = data.iloc[:, -1]
print(label)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(spectra, label, test_size = 0.3, random_state = 69, stratify = label)
```

```python
decisionTree = DecisionTreeClassifier(random_state = 69)
decisionTree.fit(X_train, Y_train)
```

```python
Y_pred = decisionTree.predict(X_test)

print(classification_report(Y_test,Y_pred))
print("Accuracy", accuracy_score(Y_test, Y_pred))
```

```python
plt.figure(figsize=(20,10))
plot_tree(decisionTree, filled = True, feature_names = spectra.columns, class_names = decisionTree.classes_, rounded = True, proportion = False)
plt.show()
```
